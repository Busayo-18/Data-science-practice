```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

PROBLEM 1: plot a line chart of numbers 1-10

```python
# Data for the line chart
x = range(1, 11)
y = range(1, 11)

# Create the line chart
plt.plot(x, y, linestyle='-',color='g',marker="o")

# Add labels and title
plt.xlabel("Numbers")
plt.ylabel("Numbers")
plt.grid(True)
plt.title("Line Chart of Numbers 1-10")

# Display the plot
plt.show()
```

PROBLEM 2: Plot a barchart of fruit vs quantity

PROBLEM 3: Create a ahistogram of 1,000 random numbers from a normal distribution

```python
# Generate 1000 random numbers from a normal distribution
data = np.random.randn(1000)

# Create the histogram
plt.hist(data, bins=50, density=True, alpha=0.7, color='green')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 Random Numbers from Normal Distribution")

# Display the plot
plt.show()
```

PROBLEM 8: Use seaborn to plot a heat map of a  5x5 correlation matrix

```python
# Generate a random 5x5 matrix
data = np.random.rand(5, 5)

# Calculate the correlation matrix
corr_matrix = np.corrcoef(data)

# Plot the heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt=".2f")

# Add title
plt.title("Heatmap of a 5x5 Correlation Matrix")

# Display the plot
plt.show()
```

PROBLEM 9: Use seaborn to plot a boxplot(whisker plot) of salaries grouped by department

```python
#Sample dataset
np.random.seed(3)
df= pd.DataFrame({
    "Department":["IT"]*5 + ["HR"]*5 + ["Finance"]*5,
    "Salary":np.concatenate([np.random.normal(100000,50000,5),np.random.normal(100000,50000,5),np.random.normal(100000,50000,5)])
})

#Create a boxplot and swarmplot
sns.boxplot(x="Department",y="Salary",data=df,palette= "Set2")
sns.swarmplot(x="Department",y="Salary",data=df,color="k",alpha=0.5)

#Display
plt.title("Boxplot and Swarmplot of Salaries by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

PROBLEM 10: Plot a line chart of Covid-19 cases over time using Pandas and Matplotlib

```python
# Nigeria-only COVID-19 dataset from OWID GitHub
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/total_cases.csv"

# Load dataset
df = pd.read_csv(url)

# Extract only Nigeria column
nigeria = df[['date', 'Nigeria']]

#Convert 'date' to datetime
nigeria['date']= pd.to_datetime(nigeria['date'])

# Plot Nigeria COVID-19 new cases
plt.figure(figsize=(10,5))
plt.plot(nigeria['date'], nigeria['Nigeria'], label="Total Cases", color="green")

#Format x-axis dates
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=3))

# Add labels and title
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.title("COVID-19 Total Cases in Nigeria")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()

plt.show()
```
