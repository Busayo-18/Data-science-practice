Problem 8: drop duplicate column in a dataset

```python
# Extract the zip file
zip_file_path = 'datasets/order.csv--unzip/brazilian-ecommerce.zip'
extract_path = 'datasets/order.csv--unzip/'
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List files in the extracted directory to find the correct CSV file name
print(os.listdir(extract_path))

# Read the CSV file into a DataFrame (update filename based on the output above)
orders_df = pd.read_csv(os.path.join(extract_path, 'olist_orders_dataset.csv'))
print(orders_df.head())
```

```python
customer_id_df = pd.read_csv(os.path.join(extract_path, 'olist_customers_dataset.csv'))
print(customer_id_df.head())
```

```python
merged_df = pd.merge(orders_df,customer_id_df, on="customer_id", how = "inner")
print(merged_df[merged_df.duplicated()])
```

Problem 9: merge two dataframes

```python
orders_items_df = pd.read_csv(os.path.join(extract_path, 'olist_order_items_dataset.csv'))
merged_orders_df= pd.merge(orders_df,orders_items_df, on="order_id", how = "outer")
print(merged_orders_df)
```
