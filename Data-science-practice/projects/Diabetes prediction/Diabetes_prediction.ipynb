Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# Loading dataset in a panda dataframe
diabetes_dataset = pd.read_csv('/content/diabetes (1).csv')
```

```python
# finding the outcome mean
diabetes_dataset.groupby('Outcome').mean()
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
# Separate features (X) and target (y)
X = diabetes_dataset.drop('Outcome', axis=1)
y = diabetes_dataset['Outcome']

# Scale the features
X_scaled = scaler.fit_transform(X)

# Create a new DataFrame with scaled features
scaled_features_df = pd.DataFrame(X_scaled, columns=X.columns)

display(scaled_features_df.head())
```

```python
X = scaled_features_df
Y = diabetes_dataset['Outcome']
```

Data Training and Testing

```python
# train_test split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
Y.shape, Y_train.shape, Y_test.shape
```

Data modelling

```python
# initialize the support vector machine classifier
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

Data Evaluation

```python
#finding the accuracy score
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy of X_train prediction:',X_train_prediction)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
#accuracy score of test data
X_test_prediction= classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Acuracy on test data is:',test_data_accuracy)
```

Data prediction


```python
input_data_new = (8,99,84,0,0,35.4,0.388,50,)

# Convert input data to a pandas DataFrame with the correct column names
input_df_new = pd.DataFrame([input_data_new], columns=X.columns)

# Standardize the input data using the fitted scaler
std_data_new = scaler.transform(input_df_new)

# Make a prediction using the trained classifier
prediction_new = classifier.predict(std_data_new)

print("Input Data:", input_data_new)
print("Scaled Input Data:", std_data_new)
print("Model Prediction:", prediction_new)

if (prediction_new[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
